```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
```

```python
test_genes = ['TERT', 'FGFR3', 'PIK3CA', 'TP53', 'ERCC2', 'RHOB', 'ERBB2', 'HRAS', 'RXRA', 'ELF3', 'CDKN1A', 'KRAS', 'KDM6A', 'AKT1', 'FBXW7', 'ERBB3', 'SF3B1', 'CTNNB1', 'BRAF', 'C3orf70', 'CREBBP', 'CDKN2A', 'NRAS']
print(len(test_genes))
```

```python
df.loc[:, 'Gene':'Impact']
```

```python
#filtering the data to see the genes of interest ie the 23 genes involved in the urine bladder cancer detection test
filtered_df = df.loc[df['Gene'].isin(['TERT', 'FGFR3', 'PIK3CA', 'TP53', 'ERCC2', 'RHOB', 'ERBB2', 'HRAS', 'RXRA', 'ELF3', 'CDKN1A', 'KRAS', 'KDM6A', 'AKT1', 'FBXW7', 'ERBB3', 'SF3B1', 'CTNNB1', 'BRAF', 'C3orf70', 'CREBBP', 'CDKN2A', 'NRAS'])]
```

```python
filtered_df
```

```python
#getting the mean Tum_VAF for each patient
filtered_df_grouped = filtered_df.groupby('BCCP_ID')
mean_tumvaf_per_patient = filtered_df_grouped['Tum_VAF'].mean()
mean_tumvaf_per_patient = mean_tumvaf_per_patient.reset_index() #resetting index to convert the grouped result back into a dataframe
print(mean_tumvaf_per_patient)
```

```python
import matplotlib as plt
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize= (10, 5))
plt.bar(mean_tumvaf_per_patient['BCCP_ID'], mean_tumvaf_per_patient['Tum_VAF'])
plt.xlabel('Patient ID (BCCP ID)')
plt.ylabel('Mean Tumour Variant Allele Frequency (VAF)')
plt.title('Mean Tumour Variant Allele Frequency(VAF) Per Patient')
plt.xticks(rotation = 90)
plt.show()
```

```python
# sort the dataframe by 'Tum_VAF' in ascending order
mean_tumvaf_per_patient_sorted = mean_tumvaf_per_patient.sort_values('Tum_VAF')
#display the sorted dataframe
print(mean_tumvaf_per_patient_sorted)
```

```python
#plotting the sorted mean tumour VAF per patient
plt.figure(figsize= (10, 5))
plt.bar(mean_tumvaf_per_patient_sorted['BCCP_ID'], mean_tumvaf_per_patient_sorted['Tum_VAF'])
plt.xlabel('Patient ID (BCCP ID)')
plt.ylabel('Mean Tumour Variant Allele Frequency (VAF)')
plt.title('Mean Tumour Variant Allele Frequency(VAF) Per Patient')
plt.xticks(rotation = 90)
plt.show()
```

```python
#plotting a histogram to check for normal distribution using mean tum_VAF
plt.hist(mean_tumvaf_per_patient['Tum_VAF'], bins = 40)
plt.xlabel('Mean Tumour Variant Allele Frequency (VAF)')
plt.ylabel('Number of Patients')
plt.title('Distribution of Mean Tumour Variant Allele Frequency (VAF)')
plt.show()
```

```python
#now plotting histogram using median tum_VAF for each patient instead of mean

#first calculate median tum_VAF for each patient, then resetting index
median_tumvafs = filtered_df_grouped['Tum_VAF'].median()
median_tumvafs = median_tumvafs.reset_index()

#histogram of these patient tum_VAF medians
plt.hist(median_tumvafs['Tum_VAF'], bins = 40)
plt.xlabel('Median Tumour Variant Allele Frequency (VAF)')
plt.ylabel('Number of Patients')
plt.title('Distribution of Median Tumour Variant Allele Frequency (VAF)')
plt.show()
```

```python
#count of genes per patient
genes_per_patient = filtered_df_grouped['Gene'].count()
genes_per_patient = genes_per_patient.reset_index()

#plot of count of genes per patient
plt.figure(figsize= (10, 5))
plt.bar(genes_per_patient['BCCP_ID'], genes_per_patient['Gene'])
plt.xlabel('Patient ID')
plt.ylabel('Number of Genes')
plt.title('Number of Genes per Patient')
plt.xticks(rotation = 90)
plt.show()
```

```python
#histogram of the number of genes per patient
plt.hist(genes_per_patient['Gene'], bins = 'auto')
plt.xlabel('Number of Patients')
plt.ylabel('Count of Genes')
plt.title('Distribution of the Count of Genes')
plt.show()
```

```python
#looking at patients with only one gene mutation and identifying what genes those are

#filter for patients with only one gene mutation
one_gene = genes_per_patient.loc[genes_per_patient['Gene'] == 1]
print(one_gene)
```

```python
#instead of VAF do the total no of mutations
#total no of rows per patient id
#eg some patients with many mutations vs some with v low amt of mutations, histogram of mutation count
#histogram of total mutation count 
#group original unfiltered dataset by BCCP ID
df_grouped = df.groupby('BCCP_ID')

total_genes = df_grouped['Gene'].count()
total_genes = total_genes.reset_index()
print(total_genes)
```

```python
#graph of total mutation count per patient
plt.bar(total_genes['BCCP_ID'], total_genes['Gene'])
plt.xlabel('BCCP ID')
plt.ylabel('Total Number of Mutations')
plt.title('Total Number of Mutations Per Patient')
plt.xticks(rotation= 90)
plt.show()
```

```python
#histogram of total mutation count
plt.hist(total_genes['Gene'], bins= 20)
plt.xlabel('Number of Patients')
plt.ylabel('Total Count of mutations')
plt.title('Distribution of Total Number of Mutations Across All Patients')
plt.xticks(rotation = 90)
plt.show()
```
